```python
import pandas as pd
import mysql.connector
import re
```

```python
df = pd.read_csv('RegionData.csv')
pd.set_option('display.float_format', '{:.2f}'.format)
df = df.dropna()
df.info()
```

```python
df.rename(columns={df.columns[4]: 'Population'}, inplace=True)
df.rename(columns={df.columns[7]: 'Popltn_Largest_City_Urban_Pop'}, inplace=True)
```

```python
df['GDP, PPP (current international $)'] = df['GDP, PPP (current international $)'].apply(lambda x: re.sub('[^0-9.]', '', str(x)))
df['Population'] = df['Population'].apply(lambda x: re.sub('[^0-9.]', '', str(x)))
df['Population CGR 1960-2015'] = df['Population CGR 1960-2015'].apply(lambda x: re.sub('[^0-9.]', '', str(x)))
df['Popltn_Largest_City_Urban_Pop'] = df['Popltn_Largest_City_Urban_Pop'].apply(lambda x: re.sub('[^0-9.]', '', str(x)))
cols_to_convert = ['GDP, PPP (current international $)','Population', 'Population CGR 1960-2015','Popltn_Largest_City_Urban_Pop']
df[cols_to_convert] = df[cols_to_convert].astype(float)
df
```

```python
connection = mysql.connector.connect(
    host='localhost',
    port=3306,
    user='root',
    password='***************',
    database='worldbank'
)
```

```python
# Specify the table name in the database
table_name = 'region_data1'

# Create a cursor object
cursor = connection.cursor()

# Create the table in the database (if it doesn't exist)
create_table_query = """
CREATE TABLE IF NOT EXISTS {table_name} (
    Country_Name VARCHAR(255),
    Region_Code VARCHAR(255),
    Country_Code VARCHAR(255),
    GDP_PPP DECIMAL(18, 2),
    Population INT,
    Population_CGR FLOAT,
    Internet_Users FLOAT,
    Popltn_Largest_City_Urban_Pop DECIMAL(18, 2),
    Life_Expectancy DECIMAL(18, 2),
    Literacy_Rate DECIMAL(18, 2),
    Exports DECIMAL(18, 2)
);
""".format(table_name=table_name)
cursor.execute(create_table_query)
```

```python
# Insert the DataFrame into the MySQL database
for _, row in df.iterrows():
    insert_query = """
    INSERT INTO {table_name} (
        Country_Name, Region_Code, Country_Code, GDP_PPP, Population,
        Population_CGR, Internet_Users, Popltn_Largest_City_Urban_Pop,
        Life_Expectancy, Literacy_Rate, Exports
    ) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
    """.format(table_name=table_name)
for _, row in df.iterrows():
    values = (
        row['Country Name'], row['Region Code'], row['Country Code'],
        row['GDP, PPP (current international $)'], row['Population'],
        row['Population CGR 1960-2015'], row['Internet users (per 100 people)'],
        row['Popltn_Largest_City_Urban_Pop'],
        row['2014 Life expectancy at birth, total (years)'],
        row['Literacy rate, adult female (% of females ages 15 and above)'],
        row['Exports of goods and services (% of GDP)']
    )
    cursor.execute(insert_query, values)

# Commit the changes and close the connection
connection.commit()
connection.close()
```
